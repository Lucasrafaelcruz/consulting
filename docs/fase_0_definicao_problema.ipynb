# Definição do Problema — Churn de Clientes

## **1. Contexto de Negócio**

**Existe um problema de churn que gera perda de receita.**

Fonte dos dados:  
IBM Sample Data Sets — Telco Customer Churn  
https://www.kaggle.com/datasets/blastchar/telco-customer-churn

Através deste dataset, será realizada a predição do comportamento de churn dos clientes, com o objetivo de analisar os dados disponíveis e identificar quais fatores são mais relevantes para a retenção.

Este conjunto de dados será utilizado para fins de estudo e aplicação de conceitos de algoritmos de classificação em um contexto de negócio realista.

---

## **2. Problema a Ser Resolvido**

É necessário entender quais informações são mais relevantes para identificar clientes que apresentam churn (clientes que cancelam o uso do produto em um determinado período).

O problema torna-se recorrente quando a empresa não aproveita sua lucratividade operacional devido à ausência de uma estratégia específica para diferentes perfis de clientes.

Dessa forma, busca-se identificar clientes com alto risco de churn antes do cancelamento, possibilitando a aplicação de estratégias de retenção, como ações de remarketing e ofertas promocionais direcionadas.

---

## **3. Objetivo da Análise**

Ordenar os clientes por risco de churn com base em dados históricos.

Apoiar a tomada de decisão operacional com o objetivo de potencializar a lucratividade, ajustando a estratégia de retenção de acordo com o perfil de cada cliente.

---

## **4. Decisão de Negócio Associada ao Modelo**

O modelo será utilizado para priorizar clientes com maior risco de churn, permitindo que a empresa aloque recursos de retenção apenas para aqueles com maior probabilidade de cancelamento, otimizando o uso dos recursos dos times comercial e operacional.

---

## **5. O que é considerado um erro?**

**Falso Positivo:**  
Aplicar uma estratégia de retenção agressiva a um cliente que não apresenta risco real de churn, reduzindo o ROI da operação.

**Falso Negativo:**  
Manter a estratégia padrão para clientes que efetivamente irão cancelar o serviço.

---

## **6. Métrica de Sucesso**

Maximizar o *recall* da classe churn, priorizando a identificação do maior número possível de clientes com risco de cancelamento, com o objetivo de potencializar a lucratividade operacional dos contratos.

---

## **7. Restrições e Premissas**

Este estudo assume que os dados históricos disponíveis representam de forma adequada o comportamento dos clientes no período analisado e que os padrões observados permanecem relativamente estáveis no curto prazo.

O modelo será construído a partir de dados estáticos, não considerando mudanças comportamentais em tempo real, eventos externos ou ações de retenção recentes que não estejam registradas no conjunto de dados.

Ressalta-se que a qualidade, consistência e representatividade dos dados impactam diretamente a capacidade do modelo em generalizar seus resultados, sendo possível que fatores relevantes para o churn não estejam plenamente capturados nas variáveis disponíveis.

---

## **8. Resultado Esperado**

O modelo deverá fornecer uma estimativa do risco de churn de cada cliente, servindo como suporte à decisão para a definição da melhor estratégia operacional e comercial de retenção.
